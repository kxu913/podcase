## 计算方法
**总分10分**
* 完成度：8分。视频长度分为8份。
* 重复观看：2分。重复观看次数，超过2次也为2分。



```python
def _score(v,d):
    x = d/8
    return round(v/x,2)
```

```python

tags ={}
feeds = {}
for i,r in df.iterrows():
    if r.tag in tags.keys():
        _c = tags[r.tag]['count']
        _s =  tags[r.tag]['score']
        __x = _score(r.view_time,r.duriation)
        if _s < __x:
            tags[r.tag]['score']= __x
        if _c<=2:
            tags[r.tag]['score'] =  tags[r.tag]['score']+1
            tags[r.tag]['count'] =_c+1        
    else:
        score = _score(r.view_time,r.duriation)
        tags[r.tag]={'count':1,'score':score}
    if r.feed in feeds.keys():
        feeds[r.feed]['count'] = feeds[r.feed]['count']+1
    else:
        feeds[r.feed] ={'count':1}
                               
```

```python
tags_score_df = pd.DataFrame(tags).T
tags_score_df
```
